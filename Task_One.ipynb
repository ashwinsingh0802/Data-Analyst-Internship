# Task One - Data Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('train.csv')
```

```python
df
```

### As we can see name column contain Firstname and Lastname so we will separate it into Firstname and Lastname columns.

```python
df['Name'].str.split(', ', expand=True)
```

```python
df[['Lastname','Firstname']]=df['Name'].str.split(', ', expand=True)
df
```

```python
# Now we wll drop our previous Name column.
```

```python
df_new=df.drop(columns="Name")
df_new
```

# We have successfully separate our Name Column in Firstname and Lastname column.

```python
# We can see in our dataset and find that ticket column is not useful for us to visualization so we will drop
# Ticket column from our dataset.
```

```python
df_new=df_new.drop(columns='Ticket')
```

```python
df_new
```

### We have successfully drop our Ticket column from dataset

## Now we will check for null and missing values in our dataset

```python
df.info()
```

### As we can see there are missing values in 'Age' and 'Cabin' column so we will try to remove those column

```python
df.isnull().sum()
```

```python
# There are 177 missing values in Age,682 missing values in Cabin, and 2 in Embarked column.
```

## Now we will see percentage of missing values of each column.

```python
df_new.isnull().sum()*100/df_new.shape[0]
```

#### There are 19% missing values in Age , 77% missing values in Cabin.
#### We will fill the missing values in Age column and drop the Cabin column

```python
df_new['Age']=df_new['Age'].fillna(method='ffill')
```

```python
df_new
```

```python
df_new.isnull().sum()
```

#### We have successfully fill the missing values in Age Column

#### Now we will drop Cabin column because it has more missing values.

```python
df_new=df_new.drop(columns="Cabin")
```

```python
df_new
```

```python
df_new.isnull().sum()
```

#### Now we will fill the missing values in Embarked column

```python
df_new['Embarked']=df_new['Embarked'].fillna(method="ffill")
```

```python
df_new
```

```python
df_new.isnull().sum()
```

```python
df_new[df_new.isnull().any(axis=1)]
```

#### As we can see we have successfully cleaned our Dataset. There are no null or missing values in our dataset
